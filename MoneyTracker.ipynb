```python
import pandas as pd
import os
```

```python
def open_csv():
    userandpass=pd.read_csv('DETAILS.csv')
    return userandpass
def existing_or_new():
    print('PRESS 1 FOR EXISTING USER AND 2 FOR NEW USER')
    exs_or_new=int(input())
    return exs_or_new
def user_location():
    user_loc=users.index(user_name)
    return user_loc
def username():
    print('ENTER YOUR USERNAME : ')
    user_name=input()
    return user_name
def passwrd():
    print('ENTER PASSWORD : ')
    ipass=input()
    return ipass
def password_check():
    return ipass==password[user_loc]
def password_reset():
    print('PRESS 1 TO RE ENTER YOU PASSWORD AND PRESS 2 TO RESET YOUR PASSWORD ')
    pass_sec=int(input())
    return pass_sec
def new_reset_password():
    print('ENTER YOUR NEW PASSWORD : ')
    new_pass=input()
    return new_pass
def save_csv():
    userandpass.to_csv('DETAILS.csv',index=False)
def delete_user():
    os.remove(user_pg[user_loc])
    userandpass=userandpass[userandpass['USERNAME']!=user_name]
    return userandpass
def security_question():
    print('ANSWER YOUR SECURITY QUESTION :')
    print(security_ques[user_loc])
    for x in range(3):
        sec_ans=input()
        if sec_ans==security_ans[user_loc]:
            new_pass=new_reset_password()
            password[user_loc]=new_pass
            userandpass['PASSWORD']=password
            save_csv()
            print("LOGIN SUCCESSFUL")
            main_page_enter=True
            break
        if x==2:
            print('SORRY !! YOUR ACCOUNT CANNOT BE RECOVERED')
            userandpass=delete_user()
    return main_page_enter
def new_security_ques():
    print("SELECT A SECURITY QUESTION (1,2,3 or 4) : ")
    print("1. WHAT IS YOUR PET'S NAME?")
    print("2. WHAT IS YOUR FATHER'S NAME?")
    print("3. HOW MANY ROOMS ARE THERE IN YOUR HOUSE?")
    print("4. WHICH YEAR YOU COMPLETED YOUR HIGH SCHOOL?")
    choice = 5
    while(choice>4):
        choice = int(input())
        if(choice == 1):
            sec_question = "WHAT IS YOUR PET'S NAME?"
        elif(choice == 2): 
            sec_question = "WHAT IS YOUR FATHER'S NAME?"
        elif(choice == 3):
            sec_question = "HOW MANY ROOMS ARE THERE IN YOUR HOUSE?"
        elif(choice == 4):
            sec_question = "WHICH YEAR YOU COMPLETED YOUR HIGH SCHOOL?"
    return sec_question
```
